# Ablation Analysis Plotting

Loads existing ablation CSVs and plots them without rerunning the ablation pipeline.

```python
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt

result_dir = Path("results")
good_path = result_dir / "ablation_cumulative_good.csv"
bad_path = result_dir / "ablation_cumulative_bad.csv"
iou_path = result_dir / "ablation_cumulative_iou.csv"
weight_contrib_path = result_dir / "weight_contributions.csv"

required_paths = [good_path, bad_path, iou_path]
missing = [path for path in required_paths if not path.exists()]
if missing:
    raise FileNotFoundError(
        "Missing ablation result CSVs: " + ", ".join(str(path) for path in missing)
    )

df_good = pd.read_csv(good_path)
df_bad = pd.read_csv(bad_path)
df_iou = pd.read_csv(iou_path)

display(df_good.head())
display(df_bad.head())
display(df_iou.head())
```

```python
def plot_good_bad_comparison(df_good, df_bad):
    plt.figure(figsize=(10, 6))
    plt.plot(
        df_good["percentile"] * 100,
        df_good["accuracy"] * 100,
        marker="o",
        linestyle="-",
        label="Good Neurons",
    )
    plt.plot(
        df_bad["percentile"] * 100,
        df_bad["accuracy"] * 100,
        marker="s",
        linestyle="--",
        label="Bad Neurons",
    )
    plt.title("Comparison: Good vs Bad Neuron Cumulative Ablation Effect")
    plt.xlabel("Percentile of Neurons Ablated (%)")
    plt.ylabel("Accuracy (%)")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()


plot_good_bad_comparison(df_good, df_bad)
```

```python
plt.figure(figsize=(14, 8))
plt.plot(
    df_good["percentile"] * 100,
    df_good["accuracy"] * 100,
    marker="o",
    linewidth=2.5,
    label="Good Neurons",
    color="green",
    markersize=8,
)
plt.plot(
    df_bad["percentile"] * 100,
    df_bad["accuracy"] * 100,
    marker="s",
    linewidth=2.5,
    label="Bad Neurons",
    color="red",
    markersize=8,
)
plt.plot(
    df_iou["percentile"] * 100,
    df_iou["accuracy"] * 100,
    marker="^",
    linewidth=2.5,
    label="IoU Ranked (All)",
    color="blue",
    markersize=8,
)
plt.xlabel("Percentile of Neurons Ablated (%)", fontsize=12)
plt.ylabel("Accuracy (%)", fontsize=12)
plt.title("Ablation Analysis: Good vs Bad vs IoU Ranked")
plt.legend(loc="best", fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
