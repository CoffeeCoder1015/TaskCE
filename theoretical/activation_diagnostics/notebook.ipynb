# Activation Diagnostics

Compare saved base and fine-tuned activations, with PCA available as a second view.


```python
from pathlib import Path
from theoretical.activation_diagnostics.analysis import run
from theoretical.activation_diagnostics.pca import run as run_pca
from theoretical.activation_source import (
    resolve_activation_path,
    resolve_latest_checkpoint,
)

# Configuration
model_id = "LiquidAI/LFM2.5-1.2B-Thinking"
task_name = "snli"
dataset_name = "snli"
layer = "model.layers.8.feed_forward"
checkpoint = resolve_latest_checkpoint(model_id, task_name)
base_path = resolve_activation_path(model_id, dataset_name, layer)
finetuned_path = resolve_activation_path(
    model_id,
    dataset_name,
    layer,
    checkpoint=checkpoint,
)
output_directory = Path("theoretical/activation_diagnostics/data")
alpha = 0.5
min_acts = 10
```

```python
results = run(base_path, finetuned_path, task_name=task_name, alpha=alpha, min_acts=min_acts, output_directory=output_directory)
pca_path = run_pca(task_name, finetuned_path, output_path=output_directory / f"{task_name}_pca.png")
results, pca_path
```
